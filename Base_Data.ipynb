```python


with open('RVCT_Sites.txt', 'r') as f:
  reader = csv.reader(f)
  your_list = list(reader)

print(your_list)


```

```python
import itertools

[x for x in your_list if 'TABLE' in x]
```

```python
TableCount = 0

RVCT_Headers = ['TABLE', 'BASE', 'STATE', 'COMMAND', 'REGION', 'TYPE', 'QUANTITY']
RVCT_Table = []
RVCT_Base = []
RVCT_State = []
RVCT_Command = []
RVCT_Region = []
RVCT_Type = []
RVCT_Quantity = []

Field = 0
for entry in your_list:
    try:
        if 'TABLE' in entry:
            TableCount = TableCount + 1
            print(f"Table:{TableCount}:{entry}")
            Field = 0
        else:
            RVCT_Table.append(TableCount)
            Field = Field + 1
            if Field == 0: print("Lost track!")
            if Field == 1: RVCT_Base.append(entry)
            if Field == 2: RVCT_State.append(entry)
            if Field == 3: RVCT_Command.append(entry)
            if Field == 4: RVCT_Region.append(entry)
            if Field == 5: RVCT_Type.append(entry)
            if Field == 6: RVCT_Quantity.append(entry)
            if Field == 6: Field = 0
    except:
        print("bump")

```
